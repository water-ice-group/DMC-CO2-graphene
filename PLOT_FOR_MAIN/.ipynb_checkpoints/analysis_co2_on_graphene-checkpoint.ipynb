```python
## load water + graphene data
path_to_binding_curves = '/scratch/fd385/work/co2_on_graphene/'

functionals_water_graphene = np.loadtxt(path_to_binding_curves+'water_graphene_data/water+graphene/graphene_2leg/Energy',usecols=0,dtype=str)

graphene_2leg=np.loadtxt(path_to_binding_curves+'water_graphene_data/water+graphene/graphene_2leg/Energy',usecols=1)
graphene=np.loadtxt(path_to_binding_curves+'water_graphene_data/water+graphene/graphene/Energy',usecols=1)
wgraphene_2leg=np.loadtxt(path_to_binding_curves+'water_graphene_data/water+graphene/w_2leg/Energy',usecols=1)

dmc_graphene_2leg=np.array([-99,6])
binding_graphene_2leg = 1000*(graphene_2leg-graphene-wgraphene_2leg)
delta_graphene_2leg=1000*(graphene_2leg-graphene-wgraphene_2leg)-dmc_graphene_2leg[0]
```

```python
list_of_functionals = [key for key in data.keys() if key != 'DMC']

cm = 1/2.54
fig, ax = plt.subplots(figsize=(8.9*cm, 13*cm))

# Plot for blue (Water)
y = 0
for name in list_of_functionals:
    if name in functionals_water_graphene:
        idx = np.where(functionals_water_graphene == name)[0][0]  # find the index
        dft_value = round(binding_graphene_2leg[idx])
        ax.plot(dft_value, y, marker='o', ms=7, color='navy',mfc='none',mec='navy',zorder=10, label='H$_2$O' if y == 0 else "")
    y += 1

# plot dmc value
dmc_x =  dmc_graphene_2leg[0]
dmc_err = dmc_graphene_2leg[1]
# Error bar point
ax.errorbar(x=dmc_x, y=y, xerr=dmc_err, fmt='o', mec='navy',mfc='none',capsize=6, color='navy')
# Shaded vertical region
ax.axvspan(dmc_x - dmc_err, dmc_x + dmc_err, color='navy', alpha=0.25)

# Plot for red (CO2)
y = 0
yticks = []
yticklabels = []


for name in list_of_functionals:
    curve_data = data[name]
    dft_value = round(1000 * curve_data[5, 1])
    ax.plot(dft_value, y, marker='^', ms=7, color='tab:red', mec='black',label=f'CO$_2$' if y == 0 else "")
    yticks.append(y)
    
    if(name=='R2SCAN'):
        name = 'r$^2$SCAN'
    if(name=='R2SCAN-D3'):
        name = 'r$^2$SCAN-D3'
    yticklabels.append(name)
    
    y += 1

# plot dmc value
dmc_x = 1000 * data['DMC'][1, 1]
dmc_err = 1000 * data['DMC'][1, 2]

# Error bar point
ax.errorbar(x=dmc_x, y=y, xerr=dmc_err, fmt='^', capsize=6, mec='black',color='tab:red')
# Shaded vertical region
ax.axvspan(dmc_x - dmc_err, dmc_x + dmc_err, color='tab:red', alpha=0.2)

# ticks
yticks.append(y)
yticklabels.append('DMC')
ax.set_yticks(yticks)
ax.set_yticklabels(yticklabels)

#minor ticks
ax.xaxis.set_minor_locator(AutoMinorLocator(5))  # 5 minor ticks between each major tick
ax.minorticks_on()
ax.yaxis.set_minor_locator(plt.NullLocator())

# label
ax.set_xlabel(r'Interaction energy [meV]', fontsize=12)
ax.set_xlim([-225, 8])
ax.tick_params(axis='x', labelsize=10)



plt.grid(ls='dotted')
ax.legend(ncols=1, edgecolor='black', fontsize=9, fancybox=False,bbox_to_anchor=(.51, .35))

plt.tight_layout()
plt.savefig('figure_3/figure-3.pdf')
plt.show()
```
