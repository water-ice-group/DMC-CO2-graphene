```python
list_of_functionals = [key for key in data.keys() if key != 'DMC']

cm = 1/2.54
fig, ax = plt.subplots(figsize=(8.9*cm, 13*cm))

# Plot for blue (Water)
y = 0
for name in list_of_functionals:
    if name in functionals_water_graphene:
        idx = np.where(functionals_water_graphene == name)[0][0]  # find the index
        dft_value = round(binding_graphene_2leg[idx])
        ax.plot(dft_value, y, marker='o', ms=7, color='royalblue',mec='black', label='H$_2$O' if y == 0 else "")
    y += 1

# plot dmc value
dmc_x =  dmc_graphene_2leg[0]
dmc_err = dmc_graphene_2leg[1]
# Error bar point
ax.errorbar(x=dmc_x, y=y, xerr=dmc_err, fmt='s', mec='black',capsize=6, color='royalblue')
# Shaded vertical region
ax.axvspan(dmc_x - dmc_err, dmc_x + dmc_err, color='royalblue', alpha=0.4)

# Plot for red (CO2)
y = 0
yticks = []
yticklabels = []


for name in list_of_functionals:
    curve_data = data[name]
    dft_value = round(1000 * curve_data[5, 1])
    ax.plot(dft_value, y, marker='s', ms=7,color='tab:red', mec='black',label=f'CO$_2$' if y == 0 else "")
    yticks.append(y)
    yticklabels.append(name)
    y += 1

# plot dmc value
dmc_x = 1000 * data['DMC'][1, 1]
dmc_err = 1000 * data['DMC'][1, 2]

# Error bar point
ax.errorbar(x=dmc_x, y=y, xerr=dmc_err, fmt='s', capsize=6, mec='black',color='tab:red')
# Shaded vertical region
ax.axvspan(dmc_x - dmc_err, dmc_x + dmc_err, color='tab:red', alpha=0.2)

# ticks
yticks.append(y)
yticklabels.append('DMC')
ax.set_yticks(yticks)
ax.set_yticklabels(yticklabels)

#minor ticks
ax.xaxis.set_minor_locator(AutoMinorLocator(5))  # 5 minor ticks between each major tick
ax.minorticks_on()
ax.yaxis.set_minor_locator(plt.NullLocator())

# label
ax.set_xlabel(r'Interaction energy [meV]', fontsize=12)
ax.set_xlim([-225, 8])
ax.tick_params(axis='x', labelsize=10)



plt.grid(ls='dotted')
ax.legend(ncols=1, edgecolor='black', fontsize=10, fancybox=False)

plt.tight_layout()
plt.savefig('figure_3/figure-3.pdf')
plt.show()
```

**Absolute Error DFT vs DMC**

```python
dmc_co2 = np.array([1000*data['DMC'][1][1],1000*data['DMC'][1][2]])
data_co2_diff_dft_dmc = []

for name in list_of_functionals:
    curve_data = data[name]
    dft_value = round(1000 * curve_data[5, 1])
    diff = dft_value - dmc_co2[0]
    stochastic_dmc = dmc_co2[1]
    
    data_co2_diff_dft_dmc.append(dft_value-dmc_co2[0])
    print(f"{name:20} | DFT error: {diff:6.0f} | Stocastich DMC error: {stochastic_dmc:.0f}")
```

**Error DFT vs DMC on CO2-Water**

```python
delta_dmc = dmc_co2[0] - dmc_graphene_2leg[0]
delta_dmc_error = np.sqrt( dmc_co2[1]**2 + dmc_graphene_2leg[1]**2)

data_co2_water_diff_dft_dmc = []

for name in list_of_functionals:
    curve_data = data[name]
    co2_dft_value = round(1000 * curve_data[5, 1])
    
    if name in functionals_water_graphene:
        idx = np.where(functionals_water_graphene == name)[0][0]  # find the index
        h2o_dft_value = round(binding_graphene_2leg[idx])
        
    diff = co2_dft_value - h2o_dft_value
    diff_w_dmc = diff - delta_dmc
    stochastic_dmc = delta_dmc_error
    data_co2_water_diff_dft_dmc.append(diff-delta_dmc)
                                        
    print(f"{name:20} | DFT diff: {diff:6.0f} |  DFT error: {diff_w_dmc:6.0f}  | Stocastich DMC error: {stochastic_dmc:.0f}")
```
